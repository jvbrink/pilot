# While-løkker

```Python
while <betingelse>:
     <kodeblokk>
```

````python
penger = 10000
rente = 1.035
år = 0

xliste = []
yliste = []

while penger < 20000:
    penger *= rente
    år += 1
    
    xliste.append(år)
    yliste.append(penger)
    

plot(xliste, yliste, 'o-')
axis([0, år, 0, 21000])
xlabel('Antall år')
ylabel('Kroner på konto')
title('Høyrentekonto (3.5%)')
grid()
````
